```python
import os
import folium
import pandas as pd
import geopandas as gpd
import io
from PIL import Image

cwd = "c:\\Users\\m1865\\Desktop\\DISC"
cwd_Images_Raw = cwd + "\\Sentinel-2 Images Raw"
cwd_Images_Processed = cwd + "\\Sentinel-2 Images Processed"
```

```python
df = pd.read_csv(cwd + "\\Site.csv")
df.head()
```

```python
for i in range(df.shape[0]):
    site = df.loc[i,'Site']
    lat = df.loc[i,'Latitude']
    lon = df.loc[i, 'Longitude']
    gdf = gpd.read_file(cwd_Images_Processed + "\\" + site + "\\2500m.shp")
    gdf = gdf.to_crs(epsg = 4326)
    m = folium.Map(location=(lat, lon), tiles="Esri.WorldImagery", zoom_start=14)
    sim_geo = gpd.GeoSeries(gdf.loc[0,"geometry"]).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j, style_function=lambda x: {'fillColor': '#00000000', 'color': 'Blue'})
    geo_j.add_child(folium.Popup("2500m x 2500m"))
    geo_j.add_to(m)
    gdf = gpd.read_file(cwd_Images_Processed + "\\" + site + "\\1200m.shp")
    gdf = gdf.to_crs(epsg = 4326)
    sim_geo = gpd.GeoSeries(gdf.loc[0,"geometry"]).simplify(tolerance=0.001)
    geo_i = sim_geo.to_json()
    geo_i = folium.GeoJson(data=geo_i, style_function=lambda x: {'fillColor': '#00000000', 'color': 'Blue'})
    geo_i.add_child(folium.Popup("1200m x 1200m"))
    geo_i.add_to(m)
    m.save(cwd_Images_Processed + "\\" + site + '\\map.html')
    img_data = m._to_png(5)
    img = Image.open(io.BytesIO(img_data))
    img.save(cwd_Images_Processed + "\\" + site + "\\map.png")
    print(f"{site} Generated!")
```
