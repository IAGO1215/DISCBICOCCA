```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 
cwd = "..\\..\\"
cwd_Images_Raw = cwd + "\\Sentinel-2 Images Raw"
cwd_Images_Processed = cwd + "\\Sentinel-2 Images Processed"
```

```python
df = pd.read_excel(cwd + "\\Results\\Variogram - Fitting Results.xlsx")
df.head()
```

```python
df_GoodFit = df[df['Variogram Fitting'] == 'Good']
```

```python
fig, ax = plt.subplots(figsize = (6,6))
plt.title("30m x 30m ROI All Sites")
sns.scatterplot(data=df, x="Sill 30", y="CV 30")
plt.axhline(y=0.1, color='r', linestyle='-')
plt.axvline(x=8, color='r', linestyle='-')
plt.xticks(list(plt.xticks()[0]) + [8], labels=list(plt.xticks()[1]) + ['8'])
plt.xlim(-10)
# plt.yticks(list(plt.yticks()[0]) + [0.1])
# plt.ylim(-0.01)
ax.xaxis.get_ticklabels()[-1].set_color('red')
ax.yaxis.get_ticklabels()[3].set_color('red')
```

```python
fig, ax = plt.subplots(figsize = (6,6))
plt.title("30m x 30m ROI All Sites with Good Variogram Fitting")
sns.scatterplot(data=df_GoodFit, x="Sill 30", y="CV 30")
plt.axhline(y=0.1, color='r', linestyle='-')
plt.axvline(x=8, color='r', linestyle='-')
plt.xticks(list(plt.xticks()[0]) + [8], labels=list(plt.xticks()[1]) + ['8'])
plt.xlim(-10)
# plt.yticks(list(plt.yticks()[0]) + [0.1])
# plt.ylim(-0.01)
ax.xaxis.get_ticklabels()[-1].set_color('red')
ax.yaxis.get_ticklabels()[3].set_color('red')
```
