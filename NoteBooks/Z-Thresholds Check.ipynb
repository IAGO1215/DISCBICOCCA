```python
import os
import numpy as np
import pandas as pd
import geopandas as gpd
from matplotlib import pyplot as plt
```

```python
cwd = "c:\\Users\\m1865\\Desktop\\DISC"
cwd_Images_Raw = cwd + "\\Sentinel-2 Images Raw"
cwd_Images_Processed = cwd + "\\Sentinel-2 Images Processed"
```

```python
df_Variogram = pd.read_csv(cwd + "\\Result - Variogram.csv")
df_Variogram.head()
```

```python
df_Roman = pd.read_csv(cwd + "\\Result - Roman.csv")
df_Roman.head()
```

```python
df_Variogram_30 = df_Variogram[df_Variogram["Distance"] == 30][['Site','Distance','CV','Sill']]
df_Variogram_300 = df_Variogram[df_Variogram["Distance"] == 300][['Site','Distance','CV','Sill']]
df_Variogram_900 = df_Variogram[df_Variogram["Distance"].isin([600,900])][['Site','Distance','CV','Sill']]
df_Roman_30_300 = df_Roman.loc[np.where((df_Roman['Internal Distance'] == 30) & (df_Roman['External Distance'] == 300))][["Site","Internal Distance","External Distance","Rcv","RAW Score","ST Score"]]
df_Roman_300_900 = df_Roman.loc[np.where((df_Roman['Internal Distance'] == 300) & (df_Roman['External Distance'].isin([600,900])))][["Site","Internal Distance","External Distance","Rcv","RAW Score","ST Score"]]
```

```python
soglia_Sill_30m = 7.77
soglia_Sill_300m = 61.06
soglia_Sill_900m = 108.42
soglia_CV_30m = 0.08
soglia_CV_300m = 0.2
soglia_CV_900m = 0.25

soglia_Rcv_30_300 = 1.2
soglia_Rcv_300_900 = 0.3
soglia_RAW_30_300 = 0.42
soglia_RAW_300_900 = 1.66
soglia_ST_30_300 = 0.19
soglia_ST_300_900 = 0.65
```

```python
df_Variogram_30["Sill Test"] = np.where(df_Variogram_30["Sill"] <= soglia_Sill_30m, "Pass", "Fail")
df_Variogram_30["CV Test"] = np.where(df_Variogram_30["CV"] <= soglia_CV_30m, "Pass", "Fail")
df_Variogram_30
```

```python
df_Variogram_300["Sill Test"] = np.where(df_Variogram_300["Sill"] <= soglia_Sill_300m, "Pass", "Fail")
df_Variogram_300["CV Test"] = np.where(df_Variogram_300["CV"] <= soglia_CV_300m, "Pass", "Fail")
df_Variogram_300
```

```python
df_Variogram_900["Sill Test"] = np.where(df_Variogram_900["Sill"] <= soglia_Sill_900m, "Pass", "Fail")
df_Variogram_900["CV Test"] = np.where(df_Variogram_900["CV"] <= soglia_CV_900m, "Pass", "Fail")
df_Variogram_900
```

```python
df_Roman_30_300["Rcv Test"] = np.where(df_Roman_30_300["Rcv"] <= soglia_Rcv_30_300, "Pass", "Fail")
df_Roman_30_300["RAW Test"] = np.where(df_Roman_30_300["RAW Score"] >= soglia_RAW_30_300, "Pass", "Fail")
df_Roman_30_300["ST Test"] = np.where(df_Roman_30_300["ST Score"] >= soglia_ST_30_300, "Pass", "Fail")
df_Roman_30_300
```

```python
df_Roman_300_900["Rcv Test"] = np.where(df_Roman_300_900["Rcv"] <= soglia_Rcv_300_900, "Pass", "Fail")
df_Roman_300_900["RAW Test"] = np.where(df_Roman_300_900["RAW Score"] >= soglia_RAW_300_900, "Pass", "Fail")
df_Roman_300_900["ST Test"] = np.where(df_Roman_300_900["ST Score"] >= soglia_ST_300_900, "Pass", "Fail")
df_Roman_300_900
```
