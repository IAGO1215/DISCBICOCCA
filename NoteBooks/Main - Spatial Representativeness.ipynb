```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 
cwd = "..\\..\\"
cwd_Images_Raw = cwd + "\\Sentinel-2 Images Raw"
cwd_Images_Processed = cwd + "\\Sentinel-2 Images Processed"
cwd_Images_Results = cwd + "\\Results"
```

```python
list_Site_Moved = ['ATLAS-Mohammed V','DE-Geb','FR-Bil','FR-FBn','FR-Pue','IS-YATIR','IT-BFt','IT-SR2','PORT-FOR','UK-Amo','US-Kon','US-SERC']
```

```python
pd_Vario = pd.read_excel(cwd_Images_Results + "\\HI Average MinMaxOnEachROI with Mean, CV and Sill.xlsx", header = 1)
pd_Vario.head()
```

```python
pd_Roman = pd.read_csv(cwd_Images_Results + "\\Roman.csv")
pd_Roman = pd_Roman[pd_Roman['Internal Distance']!=30].reset_index(drop = True)
pd_Roman = pd_Roman.drop(columns=['Rcv','Rse','Rst','Rsv'])
pd_Roman.head()
```

```python
pd_Fit = pd.read_excel(cwd_Images_Results + "\\Variogram - Fitting Results.xlsx")
pd_Fit_Good = pd_Fit[pd_Fit['Variogram Fitting'] == 'Good'].reset_index(drop = True)
pd_Fit_Bad = pd_Fit[pd_Fit['Variogram Fitting'] == 'Bad'].reset_index(drop = True)
```

```python
pd_Fit_Good
```

```python
pd_Fit_Bad
```

# RAW Score

```python
# Get maximum RAW Score for each site
temp = pd_Roman.groupby('Site')['RAW Score'].idxmax()
pd_Roman_Max_RAW = pd_Roman.loc[temp].drop(columns=['ST Score']).reset_index(drop = True)
pd_Roman_Max_RAW.head()
```

```python
# Compare original sites and moved sites, and filter out the one with lower score
pd_Roman_RAW_Max_Filter = pd_Roman_Max_RAW.copy()
for i in range(pd_Roman_Max_RAW.shape[0]):
    temp_Site = pd_Roman_Max_RAW['Site'][i]
    if temp_Site in list_Site_Moved:
        index_moved = pd_Roman_Max_RAW.loc[pd_Roman_Max_RAW['Site'] == temp_Site + " Moved",'RAW Score'].index.values[0]
        RAW_original = pd_Roman_Max_RAW['RAW Score'][i]
        RAW_moved = pd_Roman_Max_RAW['RAW Score'][index_moved]
        if RAW_moved > RAW_original:
            pd_Roman_RAW_Max_Filter.drop(index=(i),inplace = True)  
        else:
            pd_Roman_RAW_Max_Filter.drop(index=(index_moved),inplace = True)
pd_Roman_RAW_Max_Filter
```

```python
pd_Roman_Final_RAW = pd.merge(pd_Vario,pd_Roman_RAW_Max_Filter,on='Site')
pd_Roman_Final_RAW = pd_Roman_Final_RAW[['Site','HI 100','HI 300','HI 600','HI 900','HI Total','Internal Distance','External Distance','RAW Score']]
pd_Roman_Final_RAW.to_csv(cwd_Images_Results + "\\Roman RAW Score and HI (All).csv", index = False)
```

```python
pd_Roman_RAW_Fit = pd.merge(pd_Fit_Bad,pd_Roman_RAW_Max_Filter,on = 'Site')
pd_Roman_RAW_Fit = pd_Roman_RAW_Fit[['Site','Internal Distance','External Distance','RAW Score']]
pd_Roman_RAW_Fit.head()
```

```python
pd_Roman_Final_RAW_Fit = pd.merge(pd_Vario,pd_Roman_RAW_Fit,on='Site')
pd_Roman_Final_RAW_Fit = pd_Roman_Final_RAW_Fit[['Site','HI 100','HI 300','HI 600','HI 900','HI Total','Internal Distance','External Distance','RAW Score']]
pd_Roman_Final_RAW_Fit.to_csv(cwd_Images_Results + "\\Roman RAW Score and HI (Bad Fit).csv", index = False)
pd_Roman_Final_RAW_Fit
```
